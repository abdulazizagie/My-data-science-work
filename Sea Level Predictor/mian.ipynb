```python
import pandas as pd
from scipy.stats import linregress
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("epa-sea-level (1).csv")
df.head()
```

```python
def draw_plot():
    plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'],c=df['CSIRO Adjusted Sea Level'])
    line1 = linregress(df['Year'],df['CSIRO Adjusted Sea Level'])

    x1 = np.arange(df['Year'].mean(),2050,1)
    y1 = line1.slope*x1 + line1.intercept

    plt.plot(x1,y1)

    df2000 = df[df['Year'] >= 2000]
    line2 = linregress(df2000['Year'],df2000['CSIRO Adjusted Sea Level'])

    x1 = np.arange(2000,2050,1)
    y1 = line2.slope*x1 + line2.intercept

    plt.plot(x1,y1)
    plt.xlabel('Year')
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sea Level')
```

```python
draw_plot()
```

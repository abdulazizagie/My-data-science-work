```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("epa-sea-level (1).csv")
print(df.head())
plt.plot(df['Year'],df['CSIRO Adjusted Sea Level'])
plt.scatter(x=df['Year'],y=df['CSIRO Adjusted Sea Level'],c=df['CSIRO Adjusted Sea Level'])
plt.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[['Year']],df[['CSIRO Adjusted Sea Level']],test_size=0.25,random_state=42)
model = LinearRegression()
model.fit(X_train,y_train)
y_predic = model.score(X_test,y_test)
plt.scatter(x=df['Year'],y=df['CSIRO Adjusted Sea Level'],c=df['CSIRO Adjusted Sea Level'])
plt.plot(df['Year'],model.predict(df[['Year']]),color='red')
```

```python
# import Thted class

from threading import Thread,current_thread

# Create a funcation containing code to be executed parallaly
def display(n,msg):
    print("ti thread detail",current_thread().ident)
    for i in  range(n):
        print(msg)

# Create new thred Here
t1 = Thread(target=display,kwargs={'n':5,'msg':"Hello world"})

# start the new thread
t1.start()

for i in range(4):
    print("welcome")
```

```python
from threading import Thread
import time
def square(num):
    print("finding square..")
    time.sleep(1)
    print(f"square of {num} is ",num)
    
def cube(num):
    print("finding cube..")
    time.sleep(1)
    print(f"cube of {num} is ",num**3)
    
begin=time.time() #epoch :- 1 jan 2000 1000
t1=Thread(target=square,args=(3,))
t2=Thread(target=cube,args=(3,))
t1.start()
t2.start()
t1.join()
t2.join()
print("total time taken:",time.time()-begin)
```

```python
import pandas as pd


def calculate_demographic_data():
    # In this challenge you must analyze demographic data using Pandas.
    # You are given a dataset of demographic data that was extracted from the 1994 Census database.
    # Here is a sample of what the data looks like:
    df = pd.read_csv("adult.data.csv")
    
    # How many people of each race are represented in this dataset?
    # This should be a Pandas series with race names as the index labels.
    # (race column)
    race_counts = df['race'].value_counts()
    print("race",race_counts)


    # What is the average age of men?
    Average_age_men = df[df['sex'] == 'Male']['age'].mean()
    print("Average_age_men",Average_age_men)


    # # What is the percentage of people who have a Bachelor's degree?
    df['education'].value_counts()
    BCS_info = df[df['education'] == 'Bachelors']
    BCS_percentage = (len(BCS_info)/len(df) )* 100
    print("Bachelor's Degree",BCS_percentage)


    # What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?
    advance_education = df[df['education'].isin(['Bachelors','Masters','Doctorate'])]
    # print(advance_education)
    percentage = (advance_education['salary'] == '>50K').mean()*100
    print("Advance education",percentage)

    # What percentage of people without advanced education make more than 50K?
    without_advance = df[df['education'].isin(['HS-grad','Some-college','Assoc-voc','11th','7th-8th','646Prof-school','9th','12th','Doctorate','5th-6th','1st-4th','Preschool'])]
    # # print(without_advance)
    percentage_withoud = (without_advance['salary'] == '>50K').mean()*100
    print("percentage_without_advance: ",percentage_withoud)

    # What is the minimum number of hours a person works per week?
    minimum = df['hours-per-week'].min()
    print("Minimum hours works is: ",minimum)


    mini = df[df['hours-per-week'].isin([1])]
    # What percentage of the people who work the minimum number of hours per week
    # have a salary of more than 50K?
    per = (mini['salary'] == '>50K').mean() * 100
    print("minimum_>50K is:",per)


    # What country has the highest percentage of people that earn >50K?
    highest_country = (
        df[["salary", "native-country"]].groupby("native-country").apply(lambda g: g.loc[g["salary"] == ">50K"].size / g.size * 100)
    )
    highest_earning_country = highest_country.idxmax()
    highest_earning_country_percentage = highest_country.max()
    print("highest percentage of people that earn >50K :",highest_earning_country," : ",highest_earning_country_percentage)

    # Identify the most popular occupation for those who earn >50K in India.
    most_popular_occupation_india =df[df['native-country'].isin(['India'])]
    eran_mony = (most_popular_occupation_india['salary'] >= '>50K').mean() *100
    print("most_popular occupation india",eran_mony)

calculate_demographic_data()
```

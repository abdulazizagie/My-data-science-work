```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("salaries.csv")
df.head()
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
df['company_n'] = le_company.fit_transform(df['company'])
df['job_n'] = le_job.fit_transform(df['job'])
df['degree_n'] = le_degree.fit_transform(df['degree'])
```

```python
df
```

```python
df_n = df.drop(['company','job','degree'],axis='columns')
df_n
```

```python
X = df_n.drop('salary_more_then_100k',axis='columns')
X
```

```python
y = df['salary_more_then_100k']
y
```

```python
from sklearn import tree
model = DecisionTreeClassifier()
model.fit(X,y)
tree.plot_tree(model)
```

```python
model.score(X,y)
```

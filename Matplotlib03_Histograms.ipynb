```python
import matplotlib.pyplot as plt
```

```python
blood_sugar_man = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women= [113,85,90,150,149,88,93,115,135,80,77,82,129,100]

plt.xlabel('Sugar range')
plt.ylabel('Total no og patients')
plt.title("Blood sugar analysis")
plt.hist([blood_sugar_man,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','red'],
         label=['men','women'],orientation='horizontal')
plt.legend()
```

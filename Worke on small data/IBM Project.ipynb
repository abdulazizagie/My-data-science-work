```python
# import library requirment
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IBM Attrition Data.csv')
```

```python
# user define funcation 
def Education(num):
    if (num==1):
        return "Below Colleg"
    elif (num==2):
        return "Colleg"
    elif (num==3):
        return "Bachelor"
    elif(num==4):
        return "Master"
    elif(num==5):
        return "Doctor"

df['Education'] = df['Education'].apply(Education)
```

```python
df.head()
```

```python
def EnvironmentSatisfaction(num):
    if (num==1):
        return 'Low'
    elif (num==2):
        return 'Medium'
    elif (num==3):
        return 'High'
    elif (num==4):
        return 'Very High'

df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].apply(EnvironmentSatisfaction)
```

```python
df.head()
```

```python
def JobSatisfaction(num):
    if(num==1):
        return 'Low'
    elif(num==2):
        return 'Medium'
    elif (num==3):
        return 'high'
    elif (num==4):
        return 'Very High'

df['JobSatisfaction']=df['JobSatisfaction'].apply(JobSatisfaction)
```

```python
df.head()
```

```python
def WorkLifeBalance(num):
    if (num==1):
        return 'Bad'
    elif(num==2):
        return 'Good'
    elif(num==3):
        return 'Better'
    elif(num==4):
        return 'Best'

df['WorkLifeBalance'] = df['WorkLifeBalance'].apply(WorkLifeBalance)
```

```python
y=df['Attrition']
```

```python
y.head()
```

```python
from patsy import dmatrices

y, x = dmatrices('Attrition ~ Age + Department + \
                  DistanceFromHome + Education + EducationField + YearsAtCompany',
                  df, return_type="dataframe")
print (x.columns)
```

```python
import numpy as np
y = np.ravel(y)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model = model.fit(x, y)

# check the accuracy on the training set
model.score(x, y)
```

```python
y.mean()
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(x,y, test_size=0.3, random_state=0)
model2=LogisticRegression()
model2.fit(X_train, y_train)
```

```python
predicted= model2.predict(X_test)
print (predicted)
```

```python
probs = model2.predict_proba(X_test)
print (probs)
```

```python
from sklearn import metrics

print (metrics.accuracy_score(y_test, predicted))
print (metrics.roc_auc_score(y_test, probs[:, 1]))
```

```python
print (metrics.confusion_matrix(y_test, predicted))
print (metrics.classification_report(y_test, predicted))
```

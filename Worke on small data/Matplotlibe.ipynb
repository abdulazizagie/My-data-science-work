# Step of Create Plot-Example

####
    1. import numpy for generation random number
    2. import matplotlib Library


```python
from click import style
import numpy as np
# numpy generate random numbers
import matplotlib.pyplot as plt
# Plot the number
from matplotlib import style
# set the grid stile
```

#### generate random number totle(10)


```python
randomNumber = np.random.rand(10)
```

#### view them

```python
randomNumber
```

```python
# select the style of the plo
style.use('ggplot')
# plot the random number
plt.plot(randomNumber,'g',label='line one',linewidth=2)
# axis is number of random number (index)
plt.xlabel('Range')
#y axis is actual random number
plt.ylabel('Number')
# title of the plat
plt.title('First plot')
```

```python
plt.legend()
plt.show()
```

# Matplotlibe Project 02

```python
# import requires libraryies
import matplotlib.pyplot as plt
```

```python
# Data from ohio stat keading causes of death in the state from year 2012
# cause of deaths
cause = 'Chronic Diseses','Untentional Injuries','Alzhemiers','Infueza and Pneumonla','Sepsis','other'
# percenttile
percentile = [62,5,4,2,1,26]
```

```python
# set the pie chart plot properites
# set figure size
plt.figure(figsize=(10,10))
# explode the largest pie the data set'
explode = (0.05,0,0,0,0,0)
# set pie chart propreties
plt.pie(percentile,labels=cause,explode=explode,autopct = '%1.1f%%',startangle=70)
# set axis equal to draw pie circle
plt.axis('equal')
# set title of the pie chart
plt.title('Ohio state -2012:leading Ca')
# show the plot (pie chart)
```

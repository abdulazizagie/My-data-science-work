# Pair Plot

```python
# import requirment library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# To view the plot in notbook
%matplotlib inline
```

```python
# import the auto data set
df_auto_dataset = pd.read_csv('auto-mpg.csv')
```

```python
# view the top 5 recoder
df_auto_dataset.head()
```

```python
# write a user defined funcation for origine
# 1-USA, 2-Europe ,3-Asia
def  origin(num):
    if  num==1:
        return 'UAS'
    elif num==2:
        return "Europe"
    else:
        return 'Asia'

# user apple funcation
df_auto_dataset['origin'] = df_auto_dataset['origin'].apply(origin)
```

```python
# view the data point after appling the user defined funcation to data set
df_auto_dataset.head(30)
```

```python
# draw the pair plot using sns for mpg , wight , orign ,and with hue origin,set the size to 4
# note: hue is variable indata set to map plot aspects to different colors

sns.pairplot(df_auto_dataset[['mpg','weight','origin']],hue='origin',diag_kind="hist",size=4)
```

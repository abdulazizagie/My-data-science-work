# pie plot

```python
# import requirment librar
import matplotlib.pyplot as plt
```

```python
# Data from Ohio  state leading cause of death in the satae from year 2012
#cause of deaths
cause = 'Chroinc Diseeases','Unintertional Injuries','Alzheimer','Infuenza and Pneumonia','Sepis','Others'
# percentile
# percentile
percentile = [62,5,4,2,1,26]
```

```python
# set the pie chart plot properties
# set figure size
plt.figure(figsize=(10,10))
# explode the largest pie inthe data set
explode = (0.05,0.05,0,0,0,0)
# set pie chart properties
plt.pie(percentile,labels=cause,explode=explode,autopct="%1.1f%%",startangle=70)
# set axis equal to draw pie as circle
plt.axis('equal')
#set title of the pie chart
plt.title("Ohio State - 2012 : Leading Cause of Death")
# show the plot (pie chart)
plt.show()
```
